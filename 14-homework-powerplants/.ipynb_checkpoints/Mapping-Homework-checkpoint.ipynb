```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
from shapely.geometry import Polygon
%matplotlib inline
import fiona
```

```python
# Make a map of power plants in the US, color coded by type
powerplants = pd.read_csv("powerplants.csv")

points = powerplants.apply(lambda row: Point(row.Longitude, row.Latitude), axis=1)
powerplants = gpd.GeoDataFrame(powerplants, geometry=points)
# gallery = gallery.to_crs(gallery.crs)
powerplants.crs = {'init': 'epsg:5071'}

powerplants.head()
```

```python
states = gpd.read_file("/Users/amandajames/Desktop/14-homework-powerplants/cb_2016_us_state_500k/cb_2016_us_state_500k.shp")
states.crs = {'init': 'epsg:5071'}
states.head(10)
```

```python
powerplants.plot()
```

```python
top = 51.3457868 # north lat
left = -124.7844079 # west long
right = -65.9513812 # east long
bottom =  20.7433195

bounds = [
            [
              left,
              bottom
            ],
            [
              right,
              bottom
            ],
            [
              right,
              top
            ],
            [
              left,
              top
            ],
            [
              left,
              bottom
            ]
]

# Turn the coordinates into a shapely polygon
powerplantbox = Polygon(bounds)

# Make sure our state CRS is in lat/lng by using to_crs
# You could also just have done
# states = states.to_crs({'init': 'epsg:4326'})
# and converted it permanently once instead of doing it multiple times later on
# ax = gpd.GeoDataFrame(newdf)
# newdf.crs={'init':'epsg:4326'}


powerplants = powerplants[powerplants.to_crs({'init': 'epsg:5071'}).within(powerplantbox)]
```

```python
powerplants.plot()
```

```python


top = 51.3457868 # north lat
left = -124.7844079 # west long
right = -65.9513812 # east long
bottom =  20.7433195

bounds = [
            [
              left,
              bottom
            ],
            [
              right,
              bottom
            ],
            [
              right,
              top
            ],
            [
              left,
              top
            ],
            [
              left,
              bottom
            ]
]

# Turn the coordinates into a shapely polygon
statesbox = Polygon(bounds)

# Make sure our state CRS is in lat/lng by using to_crs
# You could also just have done
# states = states.to_crs({'init': 'epsg:4326'})
# and converted it permanently once instead of doing it multiple times later on
# ax = gpd.GeoDataFrame(newdf)
# newdf.crs={'init':'epsg:4326'}


states = states[states.to_crs({'init': 'epsg:5071'}).within(statesbox)]
```

```python
ax=states.plot()
```

```python
# ax = powerplants.plot(column= 'PrimSource', figsize=(30,10), cmap='BuGn')
```

```python
plants_in_states = gpd.sjoin(states.reset_index(), powerplants, how='inner', op='intersects')
plants_in_states.head(10)
```

```python
# plants_in_states.plot()
```

```python
plants_in_states['plant_count'] = plants_in_states.apply(lambda row:powerplants.within(row.geometry).sum(), axis=1)
```

```python
plants_in_states.plot(column='plant_count',figsize=(24,24),cmap='Reds')
```
